## Uivariate Analysis

```python
numerical_summary = data[numerical_cols].describe()
```

```python
numerical_summary
```

```python
# Plot for numerical distributions and outliers
for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    sns.boxplot(y=data[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
for col in categorical_cols:
    value_counts = data[col].value_counts()
    percentages = data[col].value_counts(normalize=True) * 100
    summary_df = pd.DataFrame({'Count': value_counts, 'Percentage': percentages})
    print(summary_df)
```

```python
# Plot for categorical distribution
for col in categorical_cols:
    value_counts = data[col].value_counts()
    percentages = data[col].value_counts(normalize=True) * 100
    summary_df = pd.DataFrame({'Count': value_counts, 'Percentage': percentages})

    plt.figure(figsize=(10, 6))
    sns.countplot(y=data[col], order=data[col].value_counts().index, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

## Bivariate Analysis

```python
data.columns
```

```python
# Columns for bivariate analysis against 'Diagnosis'
analysis_cols = [
    'Gender', 'Country', 'Ethnicity', 'Family_History', 'Radiation_Exposure',
    'Iodine_Deficiency', 'Smoking', 'Obesity', 'Diabetes'
]
target_col = 'Diagnosis'

for col in analysis_cols:
    print(f"\nAnalyzing {col} vs. {target_col}")

    # 1. Stacked Bar Chart (Proportions)

    cross_tab = pd.crosstab(data[col], data[target_col])
    cross_tab_norm = cross_tab.div(cross_tab.sum(1), axis=0)

    plt.figure(figsize=(12, 7))
    cross_tab_norm.plot(kind='bar', stacked=True, ax=plt.gca(), cmap='viridis')
    plt.title(f'Proportion of {target_col} by {col}')
    plt.xlabel(col)
    plt.ylabel('Proportion')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title=target_col, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # 2. Count Plot (Absolute Counts)
    plt.figure(figsize=(12, 7))
    sns.countplot(y=col, hue=target_col, data=data, order=data[col].value_counts().index, palette='magma')
    plt.title(f'Count of {target_col} by {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.legend(title=target_col, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

```

```python
data.head()
```

```python
## Bivariate analysis with respect to Thyroid_Cancer_Risk

risk_order = ['Low', 'Medium', 'High']
data['Thyroid_Cancer_Risk'] = pd.Categorical(data['Thyroid_Cancer_Risk'], categories=risk_order, ordered=True)
```

```python
# For Gender vs. Thyroid_Cancer_Risk
cross_tab_gender = pd.crosstab(data['Gender'], data['Thyroid_Cancer_Risk'])
cross_tab_norm_gender = cross_tab_gender.div(cross_tab_gender.sum(1), axis=0)

plt.figure(figsize=(12, 7))
cross_tab_norm_gender.plot(kind='bar', stacked=True, ax=plt.gca(), cmap='viridis')
plt.title('Proportion of Thyroid Cancer Risk by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Thyroid_Cancer_Risk', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# For Gender vs. Thyroid_Cancer_Risk
plt.figure(figsize=(12, 7))
sns.countplot(y='Gender', hue='Thyroid_Cancer_Risk', data=data, order=data['Gender'].value_counts().index, palette='magma')
plt.title('Count of Thyroid Cancer Risk by Gender')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.legend(title='Thyroid_Cancer_Risk', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# For Country vs. Thyroid_Cancer_Risk - Stacked Bar Plot
cross_tab_country = pd.crosstab(data['Country'], data['Thyroid_Cancer_Risk'])
cross_tab_norm_country = cross_tab_country.div(cross_tab_country.sum(1), axis=0)

plt.figure(figsize=(12, 7))
cross_tab_norm_country.plot(kind='bar', stacked=True, ax=plt.gca(), cmap='viridis')
plt.title('Proportion of Thyroid Cancer Risk by Country')
plt.xlabel('Country')
plt.ylabel('Proportion')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Thyroid_Cancer_Risk', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


# For Country vs. Thyroid_Cancer_Risk - Count Plot
plt.figure(figsize=(12, 7))
sns.countplot(y='Country', hue='Thyroid_Cancer_Risk', data=data, order=data['Country'].value_counts().index, palette='magma')
plt.title('Count of Thyroid Cancer Risk by Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.legend(title='Thyroid_Cancer_Risk', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# For Family_History vs. Thyroid_Cancer_Risk - Stacked Bar Plot
cross_tab_family = pd.crosstab(data['Family_History'], data['Thyroid_Cancer_Risk'])
cross_tab_norm_family = cross_tab_family.div(cross_tab_family.sum(1), axis=0)

plt.figure(figsize=(12, 7))
cross_tab_norm_family.plot(kind='bar', stacked=True, ax=plt.gca(), cmap='viridis')
plt.title('Proportion of Thyroid Cancer Risk by Family_History')
plt.xlabel('Family_History')
plt.ylabel('Proportion')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Thyroid_Cancer_Risk', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# For Family_History vs. Thyroid_Cancer_Risk - Count Plot
plt.figure(figsize=(12, 7))
sns.countplot(y='Family_History', hue='Thyroid_Cancer_Risk', data=data, order=data['Family_History'].value_counts().index, palette='magma')
plt.title('Count of Thyroid Cancer Risk by Family_History')
plt.xlabel('Count')
plt.ylabel('Family_History')
plt.legend(title='Thyroid_Cancer_Risk', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
## Relationships between Numerical Variables (Correlation Heatmap) ---
plt.figure(figsize=(12, 10))
correlation_matrix = data[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.tight_layout()
plt.show()
```
